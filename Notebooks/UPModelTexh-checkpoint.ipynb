Modeling Tumor cell interaction with CART CD8
=======================================

## Package activation

```python
# Set up working environment for CoLoMoTo
%matplotlib inline
import numpy as np
#import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

import ginsim
import biolqm
import maboss

from colomoto_jupyter import tabulate # for fixpoint table display
from itertools import combinations # for iterating over sets
import matplotlib.pyplot as plt # for modifying plots
import pandas as pd # for the visualization of lists of states
```

# Dowload model files

```python
# Set up the required files
bnd_file ="../Models/UPTExh.bnd"
cfg_WT = "../Models/UPTExh.cfg"
cfg_LowCART = "../Models/UPTExhLowCART.cfg"
cfg_LowExh = "../Models/UPTExhLowExh.cfg"
upp_file = "../Models/UPTExh.upp"
```

# Running wild-type model

```python
model_WT = maboss.load(bnd_file, cfg_WT)
```

```python
uppModel_WT = maboss.UpdatePopulation(model_WT, upp_file)
```

```python
run_WT = uppModel_WT.run('WT')
```

# Plotting wilde-type model results

```python
OutTable = run_WT.get_nodes_stepwise_probability_distribution()
```

```python
figure = plt.figure(figsize = (6,5))
plt.rcParams.update({'legend.fontsize': 15})
plt.plot(OutTable['PopRatio']*OutTable['Tum'],color = 'red')
plt.plot(OutTable['PopRatio']*OutTable['TCell'],color = "orange")
plt.plot(OutTable['PopRatio']*OutTable['CART'],color = "black")
plt.plot(OutTable['PopRatio']*OutTable['scFv'],color = "blue")
plt.legend(["Tumor","T Cell","CART","scFv"])
figure.savefig("WT_CellTypes.pdf", bbox_inches="tight")
```
